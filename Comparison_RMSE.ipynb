```python
from numpy import concatenate
from matplotlib import pyplot as plt
```

```python
lstm_rmse = [9.76, 10.71, 11.15, 12.30, 12.10, 14.67]
```

```python
svr_rmse = [8.07, 6.73, 7.01, 7.38, 7.71, 8.01]
```

```python
LAGS = [1, 5, 10, 15, 20, 25]
```

```python
def plot_lstm_vs_svr(LAGS, lstm_rmse, svr_rmse):
    fig, ax = plt.subplots(figsize=(8,5))
    ax.set_title('Comparison of LSTM VS SVR', fontsize=16)
    ax.set_ylabel('RMSE', fontsize=16)
    ax.set_xlabel('Time Lag', fontsize=16)
    ax.plot(LAGS, lstm_rmse, label='LSTM', marker='x')
    ax.plot(LAGS, svr_rmse, label='SVR', marker='x')
    ax.legend()
    plt.show()
    fig.savefig('./SVR/lstm_vs_svr_saluel')
```

```python
plot_lstm_vs_svr(LAGS, lstm_rmse, svr_rmse)
```
